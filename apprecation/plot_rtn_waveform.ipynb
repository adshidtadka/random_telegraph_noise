```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri May 18 22:30:19 2018

@author: mahfuz
"""
import numpy as np
import scipy.io as sio 
#from scipy import stats

import matplotlib
import matplotlib.pyplot as plt
import pylab
import pandas as pd

# read binary data
# 
data = sio.loadmat('../chip02/load/180522_ch02v050r49d3_int234_time10000.mat')
print(data)
```

```python
# 
pylab.figure(figsize=(12, 8))
font = {'size' : 21}
matplotlib.rc('font', **font)
matplotlib.rcParams['lines.linewidth'] = 2
plt.rcParams['font.size'] = 8 #
nrows = 2
ncols = 2
sstart = 0

# 4
show_list = [ 's323', 's283', 's0', 's36']

# 
for k, v in enumerate(show_list):
    plt.subplot(nrows, ncols, k+1)
    plt.cla()
    y = data[v].flatten()
    x = np.array(range(y.size))
    minimum = min(y)
    maximum = max(y)
    print('min = ' + str(minimum) + ', max = ' + str(maximum))
    plt.plot(x, y, 'k', label=v, ms=0.1, lw=0.1)
    plt.axhline(y = minimum,  lw=0.5, ls=':')
    plt.axhline(y = maximum,  lw=0.5, ls=':')
    plt.xticks([])
    plt.legend()
    

#plt.xlabel('$f_\mathrm{max}$ [a.u.]')
#plt.ylabel('$\Delta f / f$ [%]')


#plt.legend()
#plt.grid()
# plt.show()
plt.savefig("./fuga.pdf")
plt.close()
```
