```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri May 18 22:30:19 2018

@author: mahfuz
"""
import numpy as np
import scipy.io as sio 
#from scipy import stats

import matplotlib
import matplotlib.pyplot as plt
import pylab

# read binary data
# ここを変更する
data = sio.loadmat('../chip02/load/180522_ch02v050r48d3_int83_time10000.mat')
print(data)
```

```python
# グラフのサイズを変更
pylab.figure(figsize=(12, 8))
font = {'size' : 21}
matplotlib.rc('font', **font)
matplotlib.rcParams['lines.linewidth'] = 2
plt.rcParams['font.size'] = 8 #フォントの大きさ
nrows = 2
ncols = 2
sstart = 0

# 16のスロット番号を指定
show_list = [ 's285', 's193', 's282', 's136']

# プロット
for k, v in enumerate(show_list):
    plt.subplot(nrows, ncols, k+1)
    plt.cla()
    y = data[v].flatten()
    x = np.array(range(y.size))
    minimum = min(y)
    maximum = max(y)
    print('min = ' + str(minimum) + ', max = ' + str(maximum))
    plt.plot(x, y, 'k', label=v, ms=0.1, lw=0.1)
    plt.axhline(y = minimum,  lw=0.5, ls=':')
    plt.axhline(y = maximum,  lw=0.5, ls=':')
    plt.xticks([])
    plt.legend()
    plt.ylim([75, 94])
    

#plt.xlabel('$f_\mathrm{max}$ [a.u.]')
#plt.ylabel('$\Delta f / f$ [%]')


#plt.legend()
#plt.grid()
# plt.show()
plt.savefig("./fuga.pdf")
plt.close()
```
